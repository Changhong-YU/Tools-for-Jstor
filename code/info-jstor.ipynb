```python
def info(data):
    df = pd.DataFrame(columns = Category)
    c = 0
    author = ''
    title = ''
    year = ''
    doi = ''
    for i in data:
        #print(i)
        if i[:9]=='@article{':
            c += 1
            print(c)
            dic = pd.Series({'Author': author, 'Title' : title , 'Year': year, 'Doi': doi })
            if c!=1:
                df = df.append(dic,ignore_index=True)
            author = ''
            title = ''
            year = ''
            doi = i[9:-2]
        else:
            if (i =='\n' or i =='}\n'):
                pass
            else:
                k = i.split('=')[0].strip(' ')
                v = i.split('=')[1].split('{')[1].split('}')[0]
                if k == 'author':
                    author = v
                if k == 'title':
                    title = v
                if k == 'year':
                    year = v
    dic = pd.Series({'Author': author, 'Title' : title , 'Year': year, 'Doi': doi })
    df = df.append(dic,ignore_index=True)
    return df
```

```python
for i in l:
    filename = i
    f = open(filename,encoding = "utf-8")
    data = f.readlines()[4:]
    df_i = info(data)
    df = pd.concat([df,df_i],ignore_index=True)
```
