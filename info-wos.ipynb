```python
import pandas as pd
import numpy as np
```

```python
Category = ['Author','Title','Year','Doi']
outpath = './' 
l = ['savedrecs.txt', 'savedrecs-2.txt', 'savedrecs-3.txt', 'savedrecs-4.txt']
df = pd.DataFrame(columns = Category)
#print(Category)
```

```python
def info(data):
    c = 0
    df = pd.DataFrame(columns = Category)
    for i in range(len(data)):
        if data[i]=='\n':
            c += 1
            #print(c)
            dic = pd.Series({'Author': AU, 'Title' : TI , 'Year': PY, 'Doi': DI })
            df = df.append(dic,ignore_index=True)
            AU = ''
            TI = ''
            PY = ''
            DI = ''
        else:
            t1 = data[i][:2]
            #print('t1: ',t1)
            try:
                t2 = data[i][3:-1]
                #print('t2: ',t2)
            except:
                pass
            if t1=='AU':
                AU = t2
            if t1=='TI':
                TI = t2
            if t1=='DI':
                DI = t2
            if t1=='PY':
                PY = t2
    #print("一个新的文件",len(df))
    return df
```

```python
for i in l:
    filename = i
    f = open(filename,encoding = "utf-8")
    data = f.readlines()
    df_i = info(data)
    df = pd.concat([df,df_i],ignore_index=True)
```
